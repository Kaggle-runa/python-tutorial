# 第5章 条件分岐 (if / elif / else)

このノートブックでは、条件によって処理を変える **条件分岐** を学びます。
データ分析でも「外れ値の処理」「カテゴリ分け」など至るところで使います。

## 学習目標
- `if / elif / else` の構文を書ける
- 比較演算子と論理演算子を使いこなせる
- インデント (字下げ) でブロックを表現することを理解する
- 入れ子の if 文を読み解ける


## 5.1 比較演算子

| 演算子 | 意味 |
|-|-|
| `==` | 等しい |
| `!=` | 等しくない |
| `>` `>=` | より大きい / 以上 |
| `<` `<=` | より小さい / 以下 |

`=` (代入) と `==` (比較) は別物。よくある間違いなので注意。


````python
print(10 == 10)      # True
print(10 != 10)      # False
print(10 >= 9)       # True
print('A' == 'a')    # False (大文字小文字は区別)
````

## 5.2 if / elif / else の構文

```python
if 条件1:
    処理1
elif 条件2:
    処理2
elif 条件3:
    処理3
else:
    処理4
```

- 条件は **上から順に評価** され、**最初に True になったブロックだけ** 実行される
- それ以外の `elif` や `else` は試されない
- **インデント (半角スペース 4 つが慣習)** でブロックの範囲を示す
- インデントがズレると `IndentationError`


````python
# 単純な if
x = 10
if x > 0:
    print('正の数')
````

````python
# if / else
x = -5
if x > 0:
    print('正の数')
else:
    print('負またはゼロ')
````

````python
# if / elif / else
temp = 22

if temp >= 30:
    label = '暑い'
elif temp >= 20:
    label = '快適'
elif temp >= 10:
    label = '涼しい'
else:
    label = '寒い'

print(label)
````

### 「最初に True」 だけが実行される

`elif` は前の条件が False のときだけ評価されます。
複数当てはまっても、最初にマッチしたものしか実行されません。


````python
x = 100
if x >= 0:
    print('0 以上')
elif x >= 50:        # x=100 でも到達しない (前で True になったから)
    print('50 以上')
elif x >= 100:
    print('100 以上')
````

## 5.3 論理演算子 (and / or / not)

| 演算子 | 意味 | 例 |
|-|-|-|
| `and` | 両方 True | `x > 0 and x < 10` |
| `or` | どちらかが True | `x == 0 or x == 100` |
| `not` | True/False を反転 | `not is_active` |


````python
age = 20
has_ticket = True

if age >= 18 and has_ticket:
    print('入場できます')
else:
    print('入場できません')
````

````python
x = 5
if x < 0 or x > 10:
    print('範囲外')
else:
    print('範囲内')
````

### 連鎖比較 (Python 独自)

`0 < x < 10` のような書き方ができます。
数学の不等式と同じ感覚で書ける、Python の便利な機能。


````python
x = 5
if 0 < x < 10:
    print('1桁')

# 同じ意味の書き方
if 0 < x and x < 10:
    print('同じ意味')
````

## 5.4 入れ子の if 文

if の中にさらに if を書けます。ただし読みづらくなるので、
**3 段以上のネストは避ける** のが鉄則。


````python
score = 85
attended = True

if attended:
    if score >= 80:
        print('優')
    elif score >= 60:
        print('良')
    else:
        print('可')
else:
    print('欠席')
````

````python
# and で書き直すとフラットになる
if attended and score >= 80:
    print('優')
elif attended and score >= 60:
    print('良')
elif attended:
    print('可')
else:
    print('欠席')
````

## 5.5 if 文と複合代入を組み合わせて変数を変える

データ分析の前処理では「条件によって特定の変数を上書きする」処理がよく出てきます。


````python
# 商品の割引: 1000 円以上なら 10% off
price = 1500

if price >= 1000:
    price = price * 0.9   # ← 元の変数を上書き

print(price)
````

````python
# クリッピング: 0〜100 の範囲に収める
value = 120

if value < 0:
    value = 0
elif value > 100:
    value = 100

print(value)   # 100
````

## 5.6 三項演算子 (条件式)

`値1 if 条件 else 値2` で 1 行で書ける書き方。


````python
x = 7
label = '偶数' if x % 2 == 0 else '奇数'
print(label)
````

## 5.7 練習問題

次のコードの結果を予想してから実行しましょう。


````python
# Q1. 何が表示される?
x = 75
if x >= 90:
    grade = 'A'
elif x >= 70:
    grade = 'B'
elif x >= 50:
    grade = 'C'
else:
    grade = 'D'
print(grade)
````

**Q1 解説**

- `75 >= 90` → False
- `75 >= 70` → True → `grade = 'B'`
- 出力: `B`


````python
# Q2. 何が表示される? (and と or の組み合わせ)
age = 22
member = False

if age >= 18 and member:
    fee = 800
elif age >= 18 or member:
    fee = 1200
else:
    fee = 1500

print(fee)
````

**Q2 解説**

- 1 つ目: `22 >= 18 and False` → `True and False` → False
- 2 つ目: `22 >= 18 or False` → `True or False` → True → `fee = 1200`
- 出力: `1200`


````python
# Q3. これも追って計算してみよう
price = 800
member = True

if price >= 1000:
    rate = 0.3
elif member:
    rate = 0.1
else:
    rate = 0.0

if rate > 0:
    price = price * (1 - rate)

print('{}, {}'.format(int(price), rate))
````

**Q3 解説**

1. `price = 800`, `member = True`
2. 1 つ目の if: `price >= 1000` → False
3. `elif member` → True なので `rate = 0.1`
4. 2 つ目の if: `rate > 0` → True なので `price = 800 * 0.9 = 720.0`
5. `int(720.0)` → `720`
6. 出力: `720, 0.1`


## まとめ

- `if / elif / else` で条件分岐を表現
- 条件は上から順に評価され、最初に True になったブロックだけ実行
- インデントが揃っていないと `IndentationError`
- `and`, `or`, `not` で条件を組み合わせる
- 三項演算子 `x if cond else y` で短く書ける
