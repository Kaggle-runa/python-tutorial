# 第12章 Matplotlib 入門

**Matplotlib** は Python でグラフを描く定番ライブラリです。
公式: <https://matplotlib.org/>

## 学習目標
- 折れ線・散布図・棒グラフ・ヒストグラム・円グラフを描ける
- タイトル, 軸ラベル, 凡例, グリッドを追加できる
- 複数のグラフを並べて描ける (subplot)


## 12.1 import

`pyplot` は `plt` として import するのが慣例。
日本語を使うときは `japanize-matplotlib` を入れておくと便利。


```python
import numpy as np
import matplotlib.pyplot as plt

# 日本語ラベルを文字化けさせない (要 pip install japanize-matplotlib)
try:
    import japanize_matplotlib
except ImportError:
    print('japanize-matplotlib 未インストール (日本語が□になるかもしれません)')

%matplotlib inline
```

## 12.2 折れ線グラフ (plot)

時系列やトレンドの可視化に使います。


```python
x = [1, 2, 3, 4, 5]
y = [100, 250, 380, 500, 700]

plt.plot(x, y)
plt.title('折れ線グラフ')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

```python
# 複数の線, 凡例つき
x = np.arange(0, 10, 0.1)
plt.plot(x, np.sin(x), label='sin')
plt.plot(x, np.cos(x), label='cos')
plt.title('三角関数')
plt.legend()
plt.grid(True)
plt.show()
```

## 12.3 散布図 (scatter)

2 変数の関係性 (相関) を見るのに最適。


```python
np.random.seed(0)
x = np.random.rand(100) * 10
y = x * 1.5 + np.random.randn(100) * 2

plt.scatter(x, y, alpha=0.6)
plt.title('散布図')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

## 12.4 棒グラフ (bar)

カテゴリごとの数値比較に使います。


```python
depts = ['営業', '開発', '人事', '総務']
counts = [25, 40, 8, 12]

plt.bar(depts, counts)
plt.title('部署別人数')
plt.ylabel('人数')
plt.show()
```

## 12.5 ヒストグラム (hist)

データの分布を見るのに使います。
データ分析では **必ず最初に各列のヒストグラム** を描くと言ってよいくらい大事。


```python
data = np.random.randn(1000)   # 標準正規分布

plt.hist(data, bins=30)
plt.title('ヒストグラム')
plt.xlabel('値')
plt.ylabel('度数')
plt.grid(True)
plt.show()
```

## 12.6 円グラフ (pie)

割合の可視化に使いますが、誤読されやすいので **棒グラフのほうが推奨** されることも。


```python
values = [30, 25, 20, 15, 10]
labels = ['A', 'B', 'C', 'D', 'E']

plt.pie(values, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('構成比')
plt.show()
```

## 12.7 複数グラフを並べる (subplot)

`plt.subplots(rows, cols)` でグリッド状にグラフを並べられます。


```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

axes[0].plot([1, 2, 3], [1, 4, 9])
axes[0].set_title('左: 折れ線')

axes[1].bar(['A', 'B', 'C'], [10, 20, 15])
axes[1].set_title('右: 棒グラフ')

plt.tight_layout()
plt.show()
```

## 12.8 グラフの保存

`plt.savefig('ファイル名.png')` でファイルに保存できます。


```python
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
plt.plot(x, y)
plt.title('y = x^2')
plt.savefig('/tmp/sample_plot.png', dpi=120, bbox_inches='tight')
plt.show()
print('saved to /tmp/sample_plot.png')
```

## まとめ

- `plt.plot` 折れ線, `plt.scatter` 散布図, `plt.bar` 棒, `plt.hist` ヒスト, `plt.pie` 円
- `title`, `xlabel`, `ylabel`, `legend`, `grid` で見やすく装飾
- `plt.subplots` で複数並べ, `plt.savefig` で保存
- 日本語表示は `japanize-matplotlib` が便利
