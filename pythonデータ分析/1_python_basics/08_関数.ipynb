# 第8章 関数

このノートブックでは、処理をまとめて再利用するための **関数** を学びます。
データ分析のコードを `.py` モジュールに整理する際の基本単位です。

## 学習目標
- `def` で関数を定義できる
- 引数 (位置, デフォルト, キーワード) を使い分けられる
- `return` で値を返せる
- スコープ (変数の有効範囲) を理解する
- docstring で関数の説明を書ける


## 8.1 関数の定義と呼び出し

```python
def 関数名(引数1, 引数2, ...):
    処理
    return 戻り値
```

- `def` から始めて関数名を書く
- `()` の中に引数を書く (なくてもよい)
- `:` で本体を始め、インデントを付ける
- `return` で値を返す (なくてもよい → `None` を返す)


````python
def greet(name):
    print(f'こんにちは、{name}さん！')

greet('田中')
greet('佐藤')
````

## 8.2 引数と戻り値


````python
def add(a, b):
    return a + b

result = add(3, 4)
print(result)
````

````python
# 複数の値を返すこともできる (実体はタプル)
def divmod_(a, b):
    return a // b, a % b

q, r = divmod_(7, 3)
print(q, r)
````

````python
# return がないと None が返る
def shout(msg):
    print(msg.upper())
    # return なし

result = shout('hello')
print('戻り値:', result)
````

## 8.3 デフォルト引数

引数にデフォルト値を持たせると、省略可能になります。


````python
def power(base, exp=2):
    return base ** exp

print(power(3))          # 3 ** 2 = 9
print(power(3, 3))       # 3 ** 3 = 27
print(power(base=2, exp=10))   # キーワード引数
````

### デフォルト引数の罠

**デフォルト値にミュータブル(変更可能)な値を使うのは危険**。
関数定義時に 1 回だけ評価され、呼び出しのたびに共有されてしまうため。


````python
# 悪い例
def add_item(item, lst=[]):
    lst.append(item)
    return lst

print(add_item('a'))   # ['a']
print(add_item('b'))   # ['a', 'b']  ← 前の呼び出しの影響が残る!
````

````python
# 正しい例: None をデフォルトにして関数内で新しく作る
def add_item(item, lst=None):
    if lst is None:
        lst = []
    lst.append(item)
    return lst

print(add_item('a'))   # ['a']
print(add_item('b'))   # ['b']  ← 独立している
````

## 8.4 *args と **kwargs (可変長引数)

任意の個数の引数を受け取れる仕組み。


````python
# *args: 任意個の位置引数をタプルで受け取る
def total(*nums):
    s = 0
    for n in nums:
        s += n
    return s

print(total(1, 2, 3))
print(total(1, 2, 3, 4, 5, 6))
````

````python
# **kwargs: 任意個のキーワード引数を辞書で受け取る
def show_kwargs(**kwargs):
    for k, v in kwargs.items():
        print(f'{k} = {v}')

show_kwargs(name='田中', age=25, city='東京')
````

## 8.5 スコープ (変数の有効範囲)

関数の中で定義した変数は、関数の外からは見えません (ローカル変数)。
逆に、関数の外の変数は関数の中から **読めるが書き換えると別物になる**。


````python
def f():
    x = 10           # 関数内のローカル変数
    print('関数内のx:', x)

f()

try:
    print(x)         # ← NameError (関数の外には存在しない)
except NameError as e:
    print('NameError:', e)
````

````python
# グローバル変数を読むのは OK
y = 100

def g():
    print('関数内から見た y:', y)   # 読める

g()
````

````python
# 同じ名前で代入するとローカル変数が新しく作られる
y = 100

def h():
    y = 999          # これは新しいローカル変数 (関数外には影響しない)
    print('関数内 y:', y)

h()
print('関数外 y:', y)   # 100 のまま
````

## 8.6 docstring (関数の説明)

関数の最初に `"""..."""` でドキュメントを書く慣習があります。
他の人 (将来の自分) が読んだときに、引数や戻り値が明確になります。


````python
def add(a, b):
    """2つの数値を足して返す関数。

    Args:
        a: 数値
        b: 数値

    Returns:
        a と b の合計
    """
    return a + b

print(add.__doc__)
help(add)
````

## 8.7 型ヒント (オプション、推奨)

引数と戻り値の型を **アノテーション** として書けます。
動作は変わらないが、エディタの補完や mypy などの静的チェックで役立ちます。


````python
def add(a: int, b: int) -> int:
    return a + b

# Python 3.9 以降は list[int] のように書ける
def sum_all(numbers: list) -> int:
    return sum(numbers)

print(add(3, 4))
print(sum_all([1, 2, 3]))
````

## 8.8 ラムダ式 (無名関数)

短い関数を式の中で書ける。データ分析の `df.apply(lambda x: ...)` などで頻出。


````python
square = lambda x: x ** 2
print(square(5))

# sorted の key で使う例
words = ['banana', 'pi', 'apple']
sorted_words = sorted(words, key=lambda w: len(w))
print(sorted_words)
````

## 8.9 練習問題


````python
# Q1. 出力は?
def f(x, y=10):
    return x + y

print(f(3))
print(f(3, 5))
print(f(y=20, x=1))
````

**Q1 解説**

- `f(3)` → `y` はデフォルト 10 → `3 + 10 = 13`
- `f(3, 5)` → `5 が y` → `3 + 5 = 8`
- `f(y=20, x=1)` → キーワード指定 → `1 + 20 = 21`


````python
# Q2. 出力は?
n = 0

def inc():
    n = 100        # 関数内のローカル変数 (関数外の n は変えない)
    print('内側:', n)

inc()
print('外側:', n)
````

**Q2 解説**

関数内の `n = 100` は **新しいローカル変数**。関数外の `n` は変化しない。
出力:
```
内側: 100
外側: 0
```


## まとめ

- `def 関数名(引数):` で関数を定義し、`return` で値を返す
- デフォルト引数とキーワード引数で柔軟に呼び出せる
- 関数内の変数はローカル (関数の外からは見えない)
- docstring と型ヒントで関数の意図を明確にする
- ラムダ式は短い関数を式の中で書く用途
