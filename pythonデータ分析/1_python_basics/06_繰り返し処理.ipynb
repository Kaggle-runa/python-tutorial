# 第6章 繰り返し処理 (for / while)

このノートブックでは、**ループ** を学びます。
データ分析でも、リスト全要素にアクセスしたり、特定の条件を満たすまで処理を続けたりと頻出です。

## 学習目標
- `for` で繰り返し処理を書ける
- `while` で条件付き繰り返しを書ける

- ループのトレース (1 ステップずつ追う) ができる


## 6.1 for ループの基本

`for 変数 in イテラブル:` の形で、リストや range の要素を順番に取り出します。


````python
for i in [10, 20, 30]:
    print(i)
````

### range の使い方

`range(stop)` / `range(start, stop)` / `range(start, stop, step)` の 3 通り。
`stop` の値は **含まれない** ことに注意 (スライスと同じ感覚)。


````python
# range(n): 0 から n-1 まで
for i in range(5):
    print(i, end=' ')
print()

# range(start, stop)
for i in range(3, 8):
    print(i, end=' ')
print()

# range(start, stop, step)
for i in range(0, 20, 4):
    print(i, end=' ')
print()

# 逆順 (step を負に)
for i in range(10, 0, -1):
    print(i, end=' ')
print()
````

## 6.2 while ループの基本

`while 条件:` の条件が True の間、ブロックを繰り返します。
**条件を必ずいつかは False にする** ことを意識しないと無限ループになります。


````python
x = 0
while x < 5:
    print(x)
    x = x + 1   # ←これを忘れると無限ループに!
````

````python
# while else は珍しいが知っておくと便利
# (break しなかったときだけ else が実行される)
n = 0
while n < 3:
    print(n)
    n += 1
else:
    print('完走')
````

## 6.3 continue と break

ループの中で次の 2 つを使うと、流れを途中で変えられます。

- `continue`: その回の **残りをスキップ** して **ループの先頭に戻る**
- `break`: ループから **完全に抜ける**


````python
# continue の例: 3 をスキップ
for i in range(6):
    if i == 3:
        continue
    print(i)
````

````python
# break の例: 4 で抜ける
for i in range(10):
    if i == 4:
        break
    print(i)
````

## 6.4 ループのトレース方法

複雑なループは、頭の中で追わず **表に書き出す** のが確実です。

例:
```python
a = 0
b = 0

while a < 8:
    a = a + 1
    if a % 2 == 0:
        continue
    if a > 5:
        break
    b = b + a

print('{0}, {1}'.format(a, b))
```

| 反復 | 先頭の a | a+=1 後 | a%2==0? | a>5? | b への加算 | 反復終了時 |
|-|-|-|-|-|-|-|
| 1 | 0 | 1 | No (1 は奇数) | No | +1 | a=1, b=1 |
| 2 | 1 | 2 | **Yes → continue** | - | (なし) | a=2, b=1 |
| 3 | 2 | 3 | No | No | +3 | a=3, b=4 |
| 4 | 3 | 4 | **Yes → continue** | - | (なし) | a=4, b=4 |
| 5 | 4 | 5 | No | No | +5 | a=5, b=9 |
| 6 | 5 | 6 | **Yes → continue** | - | (なし) | a=6, b=9 |
| 7 | 6 | 7 | No | **Yes → break** | (なし) | a=7, b=9 |

ループを抜けた時点で `a=7, b=9` なので、出力は `'7, 9'`。


````python
a = 0
b = 0

while a < 8:
    a = a + 1
    if a % 2 == 0:
        continue
    if a > 5:
        break
    b = b + a

print('{0}, {1}'.format(a, b))   # '7, 9'
````

### 重要な注意点: `continue` 後も条件は再評価される

`continue` で先頭に戻るとき、`while` の条件は **必ず再チェック** されます。
ループ変数を更新する前に `continue` してしまうと、無限ループになるので注意。


````python
# 危険な例 (実行しないでください、無限ループになる)
# x = 0
# while x < 5:
#     if x % 2 == 0:
#         continue        # x が更新されないままループ先頭へ → 無限ループ
#     x += 1

# 正しい例: 先に x を更新する
x = 0
while x < 5:
    x += 1
    if x % 2 == 0:
        continue
    print(x)
````

## 6.5 enumerate と zip (データ分析で頻出)

リストを使ったループの便利関数 2 つ。


````python
# enumerate: インデックスと要素を同時に取得
fruits = ['apple', 'banana', 'orange']
for i, fruit in enumerate(fruits):
    print(i, fruit)
````

````python
# zip: 複数のリストを同時にループ
names = ['田中', '佐藤', '鈴木']
ages  = [25, 30, 28]

for name, age in zip(names, ages):
    print(f'{name}さん ({age}歳)')
````

## 6.6 ネストしたループ

for の中に for を書けます。表の全マス処理や 2 重リストでよく使います。


````python
for i in range(3):
    for j in range(3):
        print(f'({i},{j})', end=' ')
    print()
````

````python
# break は **内側のループだけ** から抜ける (外側は続く)
for i in range(3):
    for j in range(3):
        if j == 2:
            break
        print(f'({i},{j})', end=' ')
    print('内ループ脱出')
````

## 6.7 練習問題


````python
# Q1. total の値は?
total = 0
for i in range(1, 6):
    total += i
print(total)
````

**Q1 解説**

`1 + 2 + 3 + 4 + 5 = 15`。


````python
# Q2. ループのトレース (continue/break 入り)
n = 0
s = 0
while n < 6:
    n = n + 1
    if n % 3 == 0:
        continue
    if n > 4:
        break
    s = s + n
print('{}, {}'.format(n, s))
````

**Q2 解説**

| 反復 | n の更新 | n%3==0? | n>4? | s への加算 |
|-|-|-|-|-|
| 1 | 1 | No | No | +1 → s=1 |
| 2 | 2 | No | No | +2 → s=3 |
| 3 | 3 | **Yes → continue** | - | (なし) |
| 4 | 4 | No | No | +4 → s=7 |
| 5 | 5 | No | **Yes → break** | (なし) |

最終: `n=5, s=7`。出力: `5, 7`。


````python
# Q3. enumerate 練習
items = ['赤', '青', '黄']
for i, c in enumerate(items, start=1):
    print(f'{i}: {c}')
````

**Q3 解説**

`enumerate(items, start=1)` のように `start` を指定すると 1 から始められる。
出力:
```
1: 赤
2: 青
3: 黄
```


## まとめ

- `for` は決まった回数, `while` は条件次第の繰り返し
- `continue` は次の反復へ, `break` はループから完全脱出
- 複雑なループは **表に書き出して 1 ステップずつ追う**
- 無限ループ防止のため、`while` ではループ変数を必ず更新する
- `enumerate`, `zip` を使うと for ループが格段に書きやすくなる
