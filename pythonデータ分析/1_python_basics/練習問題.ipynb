## 問題 1 (文字列のインデックス)

次のコードの実行結果として正しいものはどれか?

```python
word = 'Tokyo'
print(word[5])
```

- ① T
- ② o
- ③ ''
- ④ IndexError


### 解答 1

`'Tokyo'` は 5 文字。有効インデックスは `0〜4` (T=0, o=1, k=2, y=3, o=4)。
`word[5]` は **範囲外** なので `IndexError`。

**正解: ④ IndexError**

(第2章 文字列を参照)


````python
word = 'Tokyo'
try:
    print(word[5])
except IndexError as e:
    print('IndexError:', e)
````

## 問題 2 (if/elif/else と format)

次のコードの実行結果として正しいものはどれか?

```python
price = 800
member = True

if price >= 1000:
    rate = 0.3
elif member:
    rate = 0.1
else:
    rate = 0.0

if rate > 0:
    price = price * (1 - rate)

print('{0}, {1}'.format(int(price), rate))
```

- ① 800, 0.0
- ② 720, 0.1
- ③ 560, 0.3
- ④ 800, 0.1


### 解答 2

1. `price = 800`, `member = True`
2. `price >= 1000` → False
3. `elif member` → True なので `rate = 0.1`
4. `rate > 0` → True なので `price = 800 * 0.9 = 720.0`
5. `int(720.0) = 720`
6. 出力: `720, 0.1`

**正解: ② 720, 0.1**

(第5章 条件分岐 を参照)


````python
price = 800
member = True

if price >= 1000:
    rate = 0.3
elif member:
    rate = 0.1
else:
    rate = 0.0

if rate > 0:
    price = price * (1 - rate)

print('{0}, {1}'.format(int(price), rate))
````

## 問題 3 (while + continue + break)

次のコードの実行結果として正しいものはどれか?

```python
a = 0
b = 0

while a < 8:
    a = a + 1
    if a % 2 == 0:
        continue
    if a > 5:
        break
    b = b + a

print('{0}, {1}'.format(a, b))
```

- ① 8, 16
- ② 7, 9
- ③ 6, 9
- ④ 5, 9


### 解答 3

| 反復 | 先頭の a | a+=1 後 | a%2==0? | a>5? | b への加算 |
|-|-|-|-|-|-|
| 1 | 0 | 1 | No | No | +1 → b=1 |
| 2 | 1 | 2 | **Yes → continue** | - | (なし) |
| 3 | 2 | 3 | No | No | +3 → b=4 |
| 4 | 3 | 4 | **Yes → continue** | - | (なし) |
| 5 | 4 | 5 | No | No | +5 → b=9 |
| 6 | 5 | 6 | **Yes → continue** | - | (なし) |
| 7 | 6 | 7 | No | **Yes → break** | (なし) |

最終: `a=7, b=9`。出力: `'7, 9'`

**正解: ② 7, 9**

(第6章 繰り返し処理 を参照)


````python
a = 0
b = 0

while a < 8:
    a = a + 1
    if a % 2 == 0:
        continue
    if a > 5:
        break
    b = b + a

print('{0}, {1}'.format(a, b))
````

## 問題 4 (リスト内包表記への置き換え)

次のコードをリスト内包表記に置きかえた場合、正しい記述を選択してください。

```python
words = ['cat', 'dog', 'elephant', 'rabbit', 'pig']
result = []

for w in words:
    if len(w) >= 5:
        result.append(w.upper())
```

- ① `result = [w for w in words if len(w) >= 5 .upper()]`
- ② `result = [w.upper() for w in words if len(w) >= 5]`
- ③ `result = [w.upper() if len(w) >= 5 for w in words]`
- ④ `result = {w.upper() for w in words if len(w) >= 5}`


### 解答 4

リスト内包表記の構文は **`[式 for 変数 in iterable if 条件]`**。
- 式 → `w.upper()`
- for → `for w in words`
- if → `if len(w) >= 5`

並べると `[w.upper() for w in words if len(w) >= 5]`。

不正解の理由:
- ③ `if-else` ではなくフィルタの `if` は **`for` のうしろ** に置く必要がある → SyntaxError
- ④ `{}` で囲むと **set 内包表記** になりリストにならない

**正解: ②**

(第7章 リスト内包表記 を参照)


````python
words = ['cat', 'dog', 'elephant', 'rabbit', 'pig']

# 元のループ
loop_result = []
for w in words:
    if len(w) >= 5:
        loop_result.append(w.upper())

# 内包表記
lc_result = [w.upper() for w in words if len(w) >= 5]

print('ループ版    :', loop_result)
print('内包表記版  :', lc_result)
````

## 問題 5 (リストのスライスと pop)

次のコードの実行結果として正しいものはどれか?

```python
values = [1, 2, 3, 4, 5, 6, 7]
part = values[2:6]
part.pop()
part.pop(0)
print(len(part))
```

- ① 2
- ② 3
- ③ 4
- ④ 5


### 解答 5

1. `values[2:6]` はインデックス 2,3,4,5 → `[3, 4, 5, 6]` (要素 4)
2. `part.pop()` で末尾 `6` を削除 → `[3, 4, 5]` (要素 3)
3. `part.pop(0)` で先頭 `3` を削除 → `[4, 5]` (要素 2)
4. `len(part)` = **2**

**正解: ① 2**

(第3章 リスト を参照)


````python
values = [1, 2, 3, 4, 5, 6, 7]
part = values[2:6]
print('スライス直後:', part)
part.pop()
print('pop()後   :', part)
part.pop(0)
print('pop(0)後  :', part)
print('len:', len(part))
````

## 問題 6 (辞書の追加/更新/削除と len)

次のコードの実行結果として正しいものはどれか?

```python
items = {'a': 1, 'b': 2, 'c': 3}
items['a'] = 10
items['d'] = 4
del items['b']
items['e'] = 5
print(len(items))
```

- ① 3
- ② 4
- ③ 5
- ④ 6


### 解答 6

| 操作 | 種別 | キー数の変化 | 後のキー数 |
|-|-|-|-|
| 初期 `{a, b, c}` | - | - | 3 |
| `items['a'] = 10` | 更新 (既存) | 0 | 3 |
| `items['d'] = 4` | 追加 | +1 | 4 |
| `del items['b']` | 削除 | -1 | 3 |
| `items['e'] = 5` | 追加 | +1 | 4 |

**正解: ② 4**

(第4章 辞書とタプル を参照)


````python
items = {'a': 1, 'b': 2, 'c': 3}
items['a'] = 10
items['d'] = 4
del items['b']
items['e'] = 5
print(items)
print('len:', len(items))
````

## 問題 7 (クラスとメソッド)

次のコードの実行結果として正しいものはどれか?

```python
class Box:
    def __init__(self, w, h, d):
        self.w = w
        self.h = h
        self.d = d

    def volume(self):
        return self.w * self.h * self.d

b = Box(3, 4, 5)
print(b.volume())
```

- ① 12
- ② 20
- ③ 60
- ④ None


### 解答 7

1. `Box(3, 4, 5)` で `self.w=3, self.h=4, self.d=5` が設定される
2. `b.volume()` は `self.w * self.h * self.d = 3 * 4 * 5 = 60`

**正解: ③ 60**

(第9章 クラス を参照)


````python
class Box:
    def __init__(self, w, h, d):
        self.w = w
        self.h = h
        self.d = d

    def volume(self):
        return self.w * self.h * self.d

b = Box(3, 4, 5)
print(b.volume())
````

## 問題 8 (pandas.merge の関数名)

以下のコードでは、2 つの DataFrame を共通の列 `'product_id'` で結合します。
空欄 **[ 1 ]** に当てはまる適切な語句を半角英字で入力してください。

```python
import pandas as pd

df_products = pd.DataFrame({
    'product_id': [1, 2, 3],
    'name':       ['Apple', 'Banana', 'Cherry'],
})

df_stock = pd.DataFrame({
    'product_id': [1, 2, 4],
    'stock':      [100, 50, 30],
})

# 共通する product_id を持つデータのみ抽出する「内部結合」を行う
df_inner = pd.[ 1 ](df_products, df_stock, on='product_id', how='inner')
```


### 解答 8

DataFrame 同士を結合する関数は `pd.merge`。

**正解: `merge`**

(第10章 pandas 基礎 を参照)


````python
import pandas as pd

df_products = pd.DataFrame({
    'product_id': [1, 2, 3],
    'name':       ['Apple', 'Banana', 'Cherry'],
})

df_stock = pd.DataFrame({
    'product_id': [1, 2, 4],
    'stock':      [100, 50, 30],
})

df_inner = pd.merge(df_products, df_stock, on='product_id', how='inner')
print(df_inner)
````

## 問題 9 (pandas.merge の how 引数)

以下のコードは、`df_products` のデータをすべて残しつつ `df_stock` の情報を結合します。
空欄 **[ 2 ]** に当てはまる適切な語句を半角英字で入力してください。

```python
df_left = pd.merge(df_products, df_stock, on='product_id', how='[ 2 ]')
```

(`df_products` の product_id (1, 2, 3) はすべて残し、`df_stock` に存在しない 3 については stock を NaN にする)


### 解答 9

「左の DataFrame のキーをすべて残す」結合は **左外部結合** で、`how='left'`。

参考:
- 両方にあるキーのみ残す → `inner`
- 一方にしかなくても両方残す → `outer`
- 左のキーをすべて残す → `left`
- 右のキーをすべて残す → `right`

**正解: `left`**

(第10章 pandas 基礎 を参照)


````python
df_left = pd.merge(df_products, df_stock, on='product_id', how='left')
print(df_left)
````

## 問題 10 (fillna の ffill と bfill)

以下のコードを実行した際、インデックス 2 の `col1` 列の値はどうなりますか?

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'col1': [1.0, 2.0, np.nan, np.nan, 5.0],
})

result = df.fillna(method='bfill')
```

- ① NaN
- ② 2.0
- ③ 5.0
- ④ 1.0


### 解答 10

`bfill` は **後ろ (下) の有効な値で埋める** メソッド。

| index | 元の値 | bfill 後 |
|-|-|-|
| 0 | 1.0 | 1.0 |
| 1 | 2.0 | 2.0 |
| 2 | NaN | **5.0** (次の有効値) |
| 3 | NaN | **5.0** (次の有効値) |
| 4 | 5.0 | 5.0 |

**正解: ③ 5.0**

(第10章 pandas 基礎 を参照)


````python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'col1': [1.0, 2.0, np.nan, np.nan, 5.0],
})

result = df.fillna(method='bfill')
print(result)
print('\nインデックス 2 の col1:', result.loc[2, 'col1'])
````
