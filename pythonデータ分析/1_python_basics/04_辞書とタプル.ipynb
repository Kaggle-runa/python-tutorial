# 第4章 辞書とタプル・集合

このノートブックでは、リスト以外の重要なコレクション型を学びます。
特に **辞書 (dict)** はデータ分析でも頻出です (`pd.DataFrame({...})` も実は辞書から作ります)。

## 学習目標
- 辞書 (`dict`) の作成・参照・追加・更新・削除ができる
- `len()` が辞書のキー数を返すことを正確に理解する
- タプル (`tuple`) と集合 (`set`) の役割を理解する
- どんなときにどれを使うべきかを判断できる


## 4.1 辞書 (dict) の基本

辞書は **キー (key) と値 (value) のペア** をまとめたデータ型です。
`{キー: 値, ...}` で作成します。


```python
scores = {'国語': 80, '数学': 90, '英語': 75}
print(scores)
print(type(scores))
```

```python
# 空の辞書
empty = {}
print(empty)
print(type(empty))   # dict
```

## 4.2 要素の参照


```python
scores = {'国語': 80, '数学': 90, '英語': 75}

print(scores['国語'])    # 80
print(scores['数学'])    # 90
```

```python
# 存在しないキーを参照すると KeyError
try:
    print(scores['理科'])
except KeyError as e:
    print('KeyError:', e)
```

```python
# get メソッドなら、なければデフォルト値を返せる (エラーを出さない)
print(scores.get('理科'))         # None
print(scores.get('理科', 0))      # 0  (デフォルトを指定)
```

## 4.3 追加・更新・削除

**重要**: 既存キーへの代入は **更新** (個数変化なし)。
新規キーへの代入は **追加** (個数 +1)。


```python
scores = {'国語': 80, '数学': 90, '英語': 75}
print('初期:', scores, '→ len =', len(scores))

# 既存キーへの代入 → 更新 (個数は変わらない)
scores['国語'] = 95
print('更新後:', scores, '→ len =', len(scores))

# 新規キーへの代入 → 追加 (個数 +1)
scores['理科'] = 88
print('追加後:', scores, '→ len =', len(scores))

# 削除
del scores['数学']
print('削除後:', scores, '→ len =', len(scores))
```

### `len()` は何を数える?

`len(辞書)` は **キーの数** を返します (= 値の数 = ペアの数)。

整理:

| 操作 | キー数の変化 |
|-|-|
| `d['新しいキー'] = 値` | +1 (追加) |
| `d['既存キー'] = 値` | 変化なし (更新) |
| `del d['キー']` | -1 (削除) |
| `d.pop('キー')` | -1 (取り出して削除) |


## 4.4 辞書のループ処理


```python
scores = {'国語': 80, '数学': 90, '英語': 75}

# キーだけを取得 (デフォルトの挙動と同じ)
for k in scores:
    print(k)

print('-' * 20)

# 値だけを取得
for v in scores.values():
    print(v)

print('-' * 20)

# キーと値を同時に取得 (最も使う)
for k, v in scores.items():
    print(f'{k}: {v}点')
```

## 4.5 辞書のよく使うメソッド・関数


```python
scores = {'国語': 80, '数学': 90, '英語': 75}

print('キー一覧 :', list(scores.keys()))
print('値一覧   :', list(scores.values()))
print('ペア一覧 :', list(scores.items()))

# キーが存在するかの判定
print('国語' in scores)   # True
print('理科' in scores)   # False
```

## 4.6 タプル (tuple)

タプルは **変更不可 (immutable)** なリストのようなものです。`()` で作ります。


```python
point = (10, 20)
print(point[0])    # 10
print(point[1])    # 20

# タプルの長さは len() で OK
print(len(point))
```

```python
# タプルは変更しようとするとエラー
try:
    point[0] = 999
except TypeError as e:
    print('TypeError:', e)
```

### タプルの使いどころ

- **複数の値を 1 セットでまとめたい** (座標, RGB, 日付など意味のかたまり)
- **関数から複数の値を返したい** (`return a, b` は実はタプルを返している)
- 辞書のキーや集合の要素にしたい (リストはハッシュできないので使えない)


```python
# 関数から複数の値を返す例
def divmod_(a, b):
    return a // b, a % b   # 実体は (a // b, a % b) というタプル

result = divmod_(7, 3)
print(result)              # (2, 1)
print(type(result))        # tuple

# アンパック代入で受け取れる
q, r = divmod_(7, 3)
print(q, r)
```

## 4.7 集合 (set)

集合は **重複を許さない** コレクション型です。
- `{要素, ...}` で作る (空のときは `set()`)
- **順序はない** (登録順は保証されない)


```python
s = {1, 2, 3, 3, 2, 1}
print(s)              # {1, 2, 3}  ← 重複が消える
print(3 in s)         # True

# リストから重複を除く定番イディオム
xs = [1, 2, 2, 3, 3, 3, 4]
unique = list(set(xs))
print(unique)
```

### 集合演算

数学の集合演算 (和・積・差) がそのまま書けます。


```python
a = {1, 2, 3, 4}
b = {3, 4, 5, 6}

print(a | b)    # 和集合 {1, 2, 3, 4, 5, 6}
print(a & b)    # 積集合 {3, 4}
print(a - b)    # 差集合 (a にあって b にない) {1, 2}
print(a ^ b)    # 対称差 (片方だけにある) {1, 2, 5, 6}
```

## 4.8 4 つの型のまとめ

| 型 | 例 | 順序 | 変更 | 重複 | 用途 |
|-|-|-|-|-|-|
| `list` | `[1, 2, 3]` | あり | 可 | OK | 順番のあるデータ全般 |
| `tuple` | `(1, 2, 3)` | あり | 不可 | OK | 意味のかたまり, 戻り値 |
| `dict` | `{'a': 1}` | あり(挿入順) | 可 | キー不可 | 名前と値のペア |
| `set` | `{1, 2, 3}` | なし | 可 | 不可 | 重複排除, 集合演算 |


## 4.9 練習問題


```python
# Q1. 最終的な len(d) は?
d = {'a': 1, 'b': 2, 'c': 3}
d['a'] = 100      # 既存キーへの代入
d['d'] = 4        # 新規キー
d['e'] = 5        # 新規キー
print(d)
print(len(d))
```

**Q1 解説**

- 初期キー数: 3
- `d['a'] = 100` は更新なので変化なし
- `d['d'] = 4` で +1 → 4
- `d['e'] = 5` で +1 → **5**


```python
# Q2. 最終的な len(d) は?
d = {'x': 1, 'y': 2, 'z': 3}
d['x'] = 10
del d['y']
d['w'] = 4
print(d)
print(len(d))
```

**Q2 解説**

- 初期: 3
- `d['x'] = 10` 更新 → 3
- `del d['y']` -1 → 2
- `d['w'] = 4` 新規 +1 → **3**


## まとめ

- 辞書は **キーと値のペア**。`d[key] = value` で追加/更新, `del` で削除
- `len()` はキー数を返す
- 既存キーへの代入は **更新** で個数は変わらない
- タプルは変更不可、関数の戻り値で頻出
- 集合は重複を許さず, 集合演算ができる
