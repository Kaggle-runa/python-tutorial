# 第13章 Seaborn 入門

**Seaborn** は Matplotlib の上にあるラッパーで、より見栄えの良い統計グラフを少ないコードで描けます。
公式: <https://seaborn.pydata.org/>

## 学習目標
- 散布図 (`scatterplot`) を描ける
- ヒストグラム (`histplot`) を描ける
- 箱ひげ図 (`boxplot`) を描ける
- 相関行列のヒートマップ (`heatmap`) を描ける
- ペアプロット (`pairplot`) で全変数の関係を一括可視化できる


## 13.1 import

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

try:
    import japanize_matplotlib
except ImportError:
    pass

sns.set_theme(style='whitegrid')   # 全体テーマ
%matplotlib inline
```

## 13.2 サンプルデータの準備

```python
# 内蔵データセット tips (チップの額)
tips = sns.load_dataset('tips')
tips.head()
```

## 13.3 散布図 (scatterplot)

2 変数の関係を、カテゴリ変数で色分けして表現できます。


```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='time', size='size')
plt.title('会計額とチップの関係')
plt.show()
```

## 13.4 ヒストグラム (histplot)


```python
sns.histplot(data=tips, x='total_bill', bins=20, kde=True)
plt.title('会計額の分布')
plt.show()
```

## 13.5 箱ひげ図 (boxplot)

カテゴリごとの分布を比較するのに便利。


```python
sns.boxplot(data=tips, x='day', y='total_bill', hue='sex')
plt.title('曜日別の会計額')
plt.show()
```

```python
# バイオリンプロット (分布の形まで見える)
sns.violinplot(data=tips, x='day', y='total_bill')
plt.title('曜日別の会計額 (violin)')
plt.show()
```

## 13.6 ヒートマップ (heatmap)

行列の値を色で可視化。**相関行列の表示** に最もよく使われます。


```python
# 相関係数の行列
corr = tips.select_dtypes(include='number').corr()

sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm', center=0, vmin=-1, vmax=1)
plt.title('相関係数のヒートマップ')
plt.show()
```

## 13.7 ペアプロット (pairplot)

全ての変数ペアの散布図 + 対角線にヒストグラム/KDE。EDA の最初に一発で見られて便利。


```python
sns.pairplot(tips, hue='time')
plt.show()
```

## 13.8 カウントプロット (countplot)

カテゴリ列の各値の頻度を棒グラフで。


```python
sns.countplot(data=tips, x='day', hue='sex')
plt.title('曜日 × 性別の人数')
plt.show()
```
