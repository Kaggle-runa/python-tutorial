# 第2章 文字列の操作

このノートブックでは、文字列 (`str`) の扱い方を学びます。
文字列は「文字 (character) の並び」として **インデックス** や **スライス** が使えます。

## 学習目標
- 文字列のインデックス・スライスを使える
- インデックス外参照で `IndexError` が出ることを理解する
- `.format()` と f-string で文字列を組み立てられる
- 文字列の主要メソッドを使える


## 2.1 文字列とは

シングルクォート `'...'` またはダブルクォート `"..."` で囲まれたデータが文字列です。
どちらを使ってもよいですが、プロジェクト内で揃えるのが慣習です。


````python
s1 = 'Hello'
s2 = "Python"
print(s1, s2)
print(type(s1))
````

````python
# 文字列中にクォートを含めたい場合
print("It's nice")        # ダブルで囲めばシングルを直接入れられる
print('She said "Hi"')    # 逆も可
print('It\'s nice')      # エスケープシーケンス \' を使う方法もある
````

## 2.2 文字列の長さ

`len()` で文字数を取得できます。


````python
print(len('Python'))      # 6
print(len(''))            # 0  (空文字列)
print(len('こんにちは'))  # 5  (全角でも 1 文字は 1)
````

## 2.3 文字列のインデックス (添字アクセス)

文字列は **0 始まり** で各文字にアクセスできます。
**負のインデックス** を使うと末尾から数えられます。

```
 文字   :  P   y   t   h   o   n
 index  :  0   1   2   3   4   5
 (負)   : -6  -5  -4  -3  -2  -1
```


````python
word = 'Python'

print(word[0])    # 'P'  ← 先頭
print(word[1])    # 'y'
print(word[5])    # 'n'  ← 最後の文字 (len(word)-1)
print(word[-1])   # 'n'  ← 後ろから1番目
print(word[-2])   # 'o'
````

### 重要: 範囲外を指定すると IndexError

文字列の長さが `n` のとき、有効なインデックスは `0 〜 n-1`。
**それを超えると `IndexError`** が出ます。

たとえば 3 文字の文字列なら有効インデックスは 0, 1, 2 のみ。


````python
# 3 文字の文字列で範囲外アクセス
short = 'Sky'   # len = 3
print(short[0], short[1], short[2])

try:
    print(short[3])   # ← IndexError !
except IndexError as e:
    print('IndexError:', e)
````

````python
# 「インデックスが有効か」を len() と比べて確認できる
word = 'Apple'
idx = 5

if idx < len(word):
    print(word[idx])
else:
    print(f'インデックス {idx} は範囲外 (len={len(word)})')
````

## 2.4 スライス

`文字列[開始:終了]` で部分文字列を取り出せます。
**終了位置は含まれない** ことに注意。

```
 文字 : P   y   t   h   o   n
 index: 0   1   2   3   4   5
```

`word[1:4]` は **インデックス 1, 2, 3** の文字 (`y`, `t`, `h`) を取り出します。


````python
word = 'Python'

print(word[0:2])    # 'Py'    (0, 1)
print(word[2:5])    # 'tho'   (2, 3, 4)
print(word[:3])     # 'Pyt'   (省略で先頭から)
print(word[3:])     # 'hon'   (省略で末尾まで)
print(word[-3:])    # 'hon'   (後ろから 3 文字)
print(word[:])      # 'Python' (全体のコピー)
````

````python
# ステップを指定すると等間隔で取り出せる
s = 'abcdefgh'
print(s[::2])    # 'aceg'   (2 文字おき)
print(s[::-1])   # 'hgfedcba'  (逆順)
````

### 範囲外スライスは IndexError にならない

スライスの場合、範囲外を指定しても **エラーにならず、取れる分だけ返す** のがインデックスとの違い。


````python
short = 'Sky'

print(short[0:100])   # 'Sky'   (エラーにならない)
print(short[10:20])   # ''      (空文字列)
````

## 2.5 文字列の結合と繰り返し


````python
# + で結合
print('Hello' + ' ' + 'World')

# * で繰り返し
print('ab' * 3)         # 'ababab'
print('-' * 20)         # '--------------------' (区切り線によく使う)
````

## 2.6 .format() による文字列の組み立て

`'{}'` を埋め込み、`.format(値1, 値2, ...)` で順番に流し込みます。

- `{}` だけだと **左から順に** 引数が当てはまる
- `{0}`, `{1}` のように番号を指定すると順序を入れ替えられる
- `{name}` のように名前を指定するとキーワード引数で渡せる


````python
name = '田中'
score = 90

# 順番に当てはめ
print('{}さんの点数は{}点です'.format(name, score))

# 番号指定
print('{0}さんの点数は{1}点です'.format(name, score))
print('{1}, {0}'.format('A', 'B'))      # 順番を入れ替え可

# キーワード指定
print('{n}さんは{s}点'.format(n=name, s=score))
````

````python
# 桁数や書式の指定もできる
pi = 3.141592
print('{:.2f}'.format(pi))      # '3.14'    (小数2桁)
print('{:10.4f}'.format(pi))    # '    3.1416' (10桁幅、小数4桁)
print('{:>10}'.format('Hi'))    # '        Hi' (右寄せ)
print('{:<10}|'.format('Hi'))   # 'Hi        |' (左寄せ)
````

## 2.7 f-string (推奨)

Python 3.6 以降は **f-string** を使うと最も簡潔に書けます。
`f'...{変数}...'` の形で変数を直接埋め込めます。


````python
name = '田中'
age = 25
print(f'名前: {name}, 年齢: {age}')

# 式も書ける
print(f'10 + 20 = {10 + 20}')
print(f'pi を小数2桁: {3.14159:.2f}')
````

## 2.8 よく使う文字列メソッド


````python
s = '  Hello, Python!  '

print(s.strip())          # 前後の空白を除去 → 'Hello, Python!'
print(s.lower())          # 小文字
print(s.upper())          # 大文字
print(s.replace('Python', 'World'))   # 置換
print('Hello'.startswith('He'))       # True
print('Hello'.endswith('lo'))         # True
print('apple,banana,orange'.split(','))   # ['apple','banana','orange']
print(','.join(['a', 'b', 'c']))      # 'a,b,c'
````

````python
# 数値判定系
print('123'.isdigit())     # True
print('abc'.isalpha())     # True
print('abc123'.isalnum())  # True (英数のみ)
````

## 2.9 練習問題

次のコードの結果を予想してから実行しましょう。


````python
# Q1. 何が表示される?
word = 'World'
print(word[2])
````

**Q1 解説**

`'World'` のインデックスは `W=0, o=1, r=2, l=3, d=4`。
`word[2]` は `'r'`。


````python
# Q2. 何が表示される? (エラーかも)
word = 'World'
print(word[10])
````

**Q2 解説**

`'World'` は 5 文字なので有効インデックスは 0〜4。
`word[10]` は **IndexError**。


````python
word = 'World'
try:
    print(word[10])
except IndexError as e:
    print('IndexError:', e)
````

````python
# Q3. 何が表示される?
s = 'Pythonista'
print(s[2:7])
````

**Q3 解説**

スライス `[2:7]` はインデックス 2, 3, 4, 5, 6 の 5 文字。
`s = 'P y t h o n i s t a'` だから `'thoni'`。


````python
# Q4. {0}, {1} を入れ替えると?
a = 'Cat'
b = 'Dog'
print('{1} and {0}'.format(a, b))
````

**Q4 解説**

`{0}` は最初の引数 `a = 'Cat'`、`{1}` は次の引数 `b = 'Dog'`。
番号を入れ替えると `'Dog and Cat'`。


## まとめ

- 文字列はインデックスでアクセスでき、範囲外参照は `IndexError`
- スライス `[start:end]` は end を含まない (範囲外でもエラーにならない)
- `.format()` / f-string で文字列を組み立てる
- `strip`, `lower`, `replace`, `split`, `join` などのメソッドを覚えておく
