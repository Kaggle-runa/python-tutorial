# 第7章 リスト内包表記

このノートブックでは、Pythonic な書き方の代表である **リスト内包表記** を学びます。
データ分析ではリストから特定の条件を満たすものを取り出したり、
すべての要素を変換したりする処理が頻出なので、ここを習得すると一気に楽になります。

## 学習目標
- for ループと内包表記を相互変換できる
- 条件付き内包表記の構文の **順序** を間違えない
- 辞書内包表記・集合内包表記を使える
- 「式 if 条件 else 別の式」 と 「if 条件」 の違いを理解する


## 7.1 リスト内包表記の基本

通常の for ループでリストを作るコードと比較してみます。


````python
# 普通の書き方
squares = []
for x in range(5):
    squares.append(x ** 2)
print(squares)
````

````python
# 内包表記での書き方 (短くて読みやすい)
squares = [x ** 2 for x in range(5)]
print(squares)
````

### 構文

```
[ 式  for  変数  in  iterable ]
```

- **式**: 各要素から作る値
- **for 変数 in iterable**: 値の取り出し方

メリット:
1. 短く書ける
2. **新しいリストを作る意図** が一目で分かる
3. 普通の for より少しだけ高速


````python
# 文字列のリストから長さのリストを作る
words = ['apple', 'banana', 'cherry']
lengths = [len(w) for w in words]
print(lengths)
````

## 7.2 条件付き内包表記 (フィルタ)

`for` のうしろに `if` を付けると、条件を満たす要素だけを残せます。

```
[ 式  for  変数  in  iterable  if  条件 ]
```

順序がとても大事。**「式 → for → if」** の順番でしか書けません。


````python
# 偶数だけ集めて 2 倍する
even_doubles = [x * 2 for x in range(10) if x % 2 == 0]
print(even_doubles)
````

````python
# 同等の普通のループ
result = []
for x in range(10):
    if x % 2 == 0:
        result.append(x * 2)
print(result)
````

### よくある間違い

❌ `[x if x % 2 == 0 for x in range(10)]`  → **SyntaxError**
   (`if` を `for` の前に置いてしまっている。`for` の **うしろ** にフィルタを書く)

✅ `[x for x in range(10) if x % 2 == 0]`  → 正しい


````python
# 実際にエラーになることを確認
try:
    exec('x = [x if x % 2 == 0 for x in range(10)]')
except SyntaxError as e:
    print('SyntaxError:', e)
````

## 7.3 「式 if 条件 else 別の式」 (三項演算)

`if … else …` を **式の中** に書くこともできます。これは「フィルタ」ではなく
「**要素ごとに値を変える**」操作。書く位置が違うので注意。

| パターン | 用途 | 位置 |
|-|-|-|
| `[x for x in xs if cond]` | **フィルタ** (条件を満たすものだけ残す) | `for` のうしろ |
| `[x if cond else y for x in xs]` | **要素ごとに値を変える** | `for` の前 |


````python
nums = [1, 2, 3, 4, 5]

# フィルタ: 偶数だけ残す
even_only = [x for x in nums if x % 2 == 0]
print('フィルタ:', even_only)

# 値の変換: 偶数はそのまま, 奇数は -1 にする
transformed = [x if x % 2 == 0 else -1 for x in nums]
print('変換:', transformed)
````

## 7.4 for ループ → 内包表記の変換手順

機械的に変換できます。


````python
# === 元のループ ===
result = []
for w in ['apple', 'banana', 'cherry', 'date']:
    if len(w) >= 5:
        result.append(w.upper())

print(result)
````

### 変換手順

1. `append(...)` の中身 = **式**: `w.upper()`
2. `for` 行 = **for 句**: `for w in [...]`
3. `if` 行 = **if 句**: `if len(w) >= 5`

並べると:

```python
[w.upper() for w in [...] if len(w) >= 5]
```


````python
result = [w.upper() for w in ['apple', 'banana', 'cherry', 'date'] if len(w) >= 5]
print(result)
````

## 7.5 内包表記のバリエーション

`[...]` だけでなく `{...}` (辞書/集合)、`(...)` (ジェネレータ) も同じ構文で作れます。


````python
# 辞書内包表記
names = ['apple', 'banana', 'orange']
length_map = {name: len(name) for name in names}
print(length_map)
````

````python
# 集合内包表記 (重複が消える)
nums = [1, 2, 2, 3, 3, 3, 4]
squared = {x ** 2 for x in nums}
print(squared)
````

````python
# ジェネレータ式 (大きなデータでメモリ節約)
gen = (x ** 2 for x in range(1000000))   # まだ計算されていない
total = sum(gen)                          # ここで初めて 1 件ずつ計算
print(total)
````

## 7.6 ネストした内包表記 (2 重ループ)

2 重ループも内包表記で書けますが、可読性は下がりがち。


````python
# 九九表
matrix = [[i * j for j in range(1, 4)] for i in range(1, 4)]
for row in matrix:
    print(row)
````

## 7.7 練習問題


````python
# Q1. result は何になる?
nums = [1, 2, 3, 4, 5]
result = [x * 10 for x in nums if x > 2]
print(result)
````

**Q1 解説**

- フィルタ: `x > 2` を満たす要素は 3, 4, 5
- 式: `x * 10` → 30, 40, 50
- 結果: `[30, 40, 50]`


````python
# Q2. 以下の for ループを内包表記に書き換えてみよう
words = ['cat', 'elephant', 'dog', 'rabbit']

# 元のループ
result_loop = []
for w in words:
    if len(w) <= 3:
        result_loop.append(w.upper())

# ↓↓↓ ここを書く ↓↓↓
result_lc = [w.upper() for w in words if len(w) <= 3]

print('ループ版    :', result_loop)
print('内包表記版  :', result_lc)
````

**Q2 解説**

- 式: `w.upper()`
- フィルタ: `len(w) <= 3`
- 結果: `['CAT', 'DOG']`


````python
# Q3. 三項演算が入った内包表記
nums = [1, 2, 3, 4, 5]
labels = ['偶数' if x % 2 == 0 else '奇数' for x in nums]
print(labels)
````

**Q3 解説**

`x if cond else y` を **for の前** に置く形。
偶数 (2,4) は '偶数', それ以外 (1,3,5) は '奇数' になる。


````python
# Q4. 選択肢のうち、リストではないものを当ててみよう
ans1 = [x for x in range(5)]
ans2 = (x for x in range(5))       # これは何?
ans3 = {x for x in range(5)}       # これは何?
ans4 = {x: x*2 for x in range(5)}  # これは何?

for v in (ans1, ans2, ans3, ans4):
    print(type(v).__name__, v if not hasattr(v, '__next__') else '(generator)')
````

**Q4 解説**

- `[...]` → `list`
- `(...)` → `generator` (リストではない！1要素ずつ生成する)
- `{x for x in ...}` → `set`
- `{x: y for x in ...}` → `dict`


## まとめ

- 内包表記は **「式 → for → if」** の順
- フィルタの `if` は `for` の **うしろ**
- 「要素ごとに値を変える `if-else`」 は **`for` の前** (式の中)
- `{}` で set/dict, `()` で generator も同じ構文で作れる
