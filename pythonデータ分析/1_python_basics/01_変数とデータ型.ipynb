## 1.1 変数とは

変数とは、**値に名前をつけて保存しておく入れ物**のようなものです。
Python では `変数名 = 値` の形で代入できます。


```python
# 数値を代入
age = 25
print(age)
```

```python
# 文字列を代入
name = '田中'
print(name)
```

```python
# 一度代入した変数は何度でも使える
score = 80
print('現在の点数:', score)

score = score + 10   # 値の更新 (右辺で評価した結果を左辺に再代入)
print('更新後の点数:', score)
```

### 変数名のルール

- 半角アルファベット, 数字, アンダースコア `_` のみ
- **先頭は数字にできない** (`1name = ...` は NG)
- 大文字小文字は区別される (`age` と `Age` は別物)
- 予約語 (`if`, `for`, `class` など) は使えない
- 慣習として **小文字 + アンダースコア** ( `user_name` ) で書く (スネークケース)


```python
# 良い例
user_name = '田中'
total_score = 95
is_active = True

# やや悪い例 (動くけど読みにくい)
n = '田中'         # 1 文字変数は意味が伝わらない
TotalScore = 95   # キャメルケースは慣習に反する
```

## 1.2 基本のデータ型

Python で頻繁に使うデータ型は次の 4 つです。

| 型 | 説明 | 例 |
|-|-|-|
| `int` | 整数 | `10`, `-3`, `0` |
| `float` | 小数 (浮動小数点数) | `1.5`, `-0.3`, `3.14` |
| `str` | 文字列 | `'hello'`, `"Python"` |
| `bool` | 真偽値 | `True`, `False` |


```python
x = 10            # int
y = 3.14          # float
name = 'Python'   # str
is_ok = True      # bool

print(type(x), type(y), type(name), type(is_ok))
```

## 1.3 型変換 (キャスト)

数値と文字列など、異なる型の値はそのままでは一緒に扱えません。
`int()`, `float()`, `str()` を使って型を変換します。


```python
# 文字列 '10' を整数 10 に変換
num = int('10')
print(num + 5)        # 15
```

```python
# 整数 100 を文字列に変換して連結
n = 100
msg = '点数は' + str(n) + '点です'
print(msg)
```

```python
# 文字列同士を + すると連結される (足し算にはならない)
print('1' + '2')               # '12'
print(int('1') + int('2'))     # 3
```

```python
# 小数 → 整数の変換は「切り捨て」(四捨五入ではない)
print(int(3.7))    # 3
print(int(-2.5))   # -2
```

## 1.4 算術演算子

| 演算子 | 意味 | 例 | 結果 |
|-|-|-|-|
| `+` | 足し算 | `3 + 2` | 5 |
| `-` | 引き算 | `3 - 2` | 1 |
| `*` | 掛け算 | `3 * 2` | 6 |
| `/` | 割り算 (常に float を返す) | `7 / 2` | 3.5 |
| `//` | 整数除算 (商の整数部) | `7 // 2` | 3 |
| `%` | 剰余 (割った余り) | `7 % 2` | 1 |
| `**` | べき乗 | `2 ** 3` | 8 |

剰余 `%` は「ある数で割り切れるか」を判定するのに頻繁に使われます。


```python
print(7 + 2)
print(7 - 2)
print(7 * 2)
print(7 / 2)
print(7 // 2)
print(7 % 2)
print(2 ** 3)
```

### 剰余 `%` の使いどころ

ある数 `n` を `k` で割った余りが `0` なら、`n` は `k` の倍数です。
ループの中で「偶数だけ処理する」「3 の倍数だけスキップする」といったときに大活躍します。


```python
# 偶数 (2 の倍数) を判定
for n in range(1, 11):
    if n % 2 == 0:
        print(n, 'は偶数')
```

## 1.5 複合代入演算子

`x = x + 1` は `x += 1` と書けます。


```python
x = 10
x += 5    # x = x + 5 と同じ → 15
x -= 3    # x = x - 3 と同じ → 12
x *= 2    # x = x * 2 と同じ → 24
x //= 5   # x = x // 5 と同じ → 4
print(x)
```

## 1.6 練習問題

次のコードの結果を予想してから実行してみましょう。


```python
# Q1. 何が表示される?
a = 15
b = '4'
print(a + int(b) * 2)
```

**Q1 解説**

- `int(b)` は `4` (整数)
- `int(b) * 2` は `8`
- 演算子の優先順位: `*` のほうが `+` より先に評価される
- 結果: `15 + 8 = 23`


```python
# Q2. 何が表示される?
x = 17
print(x // 5, x % 5)
```

**Q2 解説**

- `17 // 5` は商 → `3`
- `17 % 5` は余り → `2`
- 出力: `3 2`
