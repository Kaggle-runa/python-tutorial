## 3.1 リストとは

リストは **順序のある複数の値をまとめた** データ型です。`[]` で作ります。


````python
scores = [85, 70, 90, 100, 60]
print(scores)
print(type(scores))
````

````python
# いろいろな型を混ぜられる (動くがおすすめはしない)
mixed = [1, 'apple', 3.14, True]
print(mixed)

# 空のリスト
empty = []
print(empty, len(empty))
````

## 3.2 要素アクセスとスライス

文字列と同じく、**0 始まり** でアクセスします。負のインデックスは末尾から。

```
要素 : 85  70  90  100  60
index:  0   1   2    3   4
 (負): -5  -4  -3   -2  -1
```


````python
scores = [85, 70, 90, 100, 60]

print(scores[0])      # 85
print(scores[-1])     # 60  (後ろから1番目)
print(scores[2])      # 90

# スライス: [start:end]  end は含まない
print(scores[1:4])    # [70, 90, 100]   ← 要素 3 個
print(scores[:3])     # [85, 70, 90]
print(scores[2:])     # [90, 100, 60]
````

### スライスの要素数の数え方

`list[a:b]` の要素数は **`b - a` 個** です (range と同じ考え方)。
インデックス `a, a+1, ..., b-1` の要素を取り出すので、合計 `b-a` 個。


````python
scores = [85, 70, 90, 100, 60]

sub = scores[1:4]
print('要素:', sub)
print('要素数:', len(sub))   # 4 - 1 = 3

sub2 = scores[0:5]
print('要素:', sub2)
print('要素数:', len(sub2))  # 5 - 0 = 5
````

## 3.3 リストの追加・更新・削除

| 操作 | メソッド | 例 |
|-|-|-|
| 末尾に追加 | `append(x)` | `lst.append(60)` |
| 任意位置に挿入 | `insert(i, x)` | `lst.insert(0, 999)` |
| 末尾を取り出して削除 | `pop()` | `lst.pop()` |
| 任意位置を取り出して削除 | `pop(i)` | `lst.pop(2)` |
| 値で削除 | `remove(x)` | `lst.remove(70)` |
| インデックスで削除 | `del lst[i]` | `del lst[0]` |
| 末尾に連結 | `extend(iter)` | `lst.extend([1,2,3])` |


````python
# append (末尾に追加 → 長さ +1)
nums = [1, 2, 3]
nums.append(4)
print(nums)         # [1, 2, 3, 4]
print(len(nums))    # 4
````

````python
# insert (任意位置に挿入 → 長さ +1)
nums = [1, 2, 3]
nums.insert(0, 999)     # インデックス 0 に挿入
print(nums)             # [999, 1, 2, 3]
````

````python
# pop (末尾を取り出して返す → 長さ -1)
nums = [10, 20, 30, 40]
last = nums.pop()
print('取り出した:', last)
print('残り:', nums)

# 引数で位置指定もできる
first = nums.pop(0)
print('取り出した:', first)
print('残り:', nums)
````

````python
# remove (値で削除)
nums = [10, 20, 30, 20]
nums.remove(20)        # 最初に見つかった 20 だけ削除
print(nums)            # [10, 30, 20]
````

## 3.4 「スライス + 操作」を順に追う練習

複合的な操作は **1 ステップずつ書き出す** と確実です。

例題:
```python
values = [1, 2, 3, 4, 5, 6]
part = values[2:5]
part.pop()
print(len(part))
```

手順:
1. `values[2:5]` は インデックス 2,3,4 の `[3, 4, 5]` (要素 3 個)
2. `part.pop()` で末尾 `5` を取り出すので `part = [3, 4]` (要素 2 個)
3. `len(part)` は **2**


````python
values = [1, 2, 3, 4, 5, 6]

part = values[2:5]
print('スライス直後:', part, '長さ', len(part))

part.pop()
print('pop後:', part, '長さ', len(part))
````

### 大切なポイント: 「スライスは新しいリスト」

`values[2:5]` で取り出した `part` は **values とは別のリスト**。
`part` を変更しても `values` には影響しません。


````python
values = [1, 2, 3, 4, 5, 6]
part = values[2:5]

part.append(999)
print('values:', values)   # 変わらない
print('part:', part)       # 999 が追加されている
````

## 3.5 リスト全体に対する関数

| 関数 | 説明 |
|-|-|
| `len(lst)` | 要素数 |
| `sum(lst)` | 合計 (数値リスト) |
| `max(lst)` | 最大値 |
| `min(lst)` | 最小値 |
| `sorted(lst)` | 昇順ソートした新リスト |
| `lst.sort()` | 元のリストを昇順ソート (返り値は None) |
| `lst.reverse()` | 元のリストを逆順に |
| `lst.count(x)` | x が登場する回数 |
| `x in lst` | x が含まれるか (True/False) |


````python
nums = [3, 1, 4, 1, 5, 9, 2, 6]

print(len(nums))
print(sum(nums))
print(max(nums))
print(min(nums))
print(sorted(nums))
print(sorted(nums, reverse=True))
print(nums.count(1))
print(3 in nums)
````

### `sorted` と `sort` の違い

| 名前 | 元のリスト | 戻り値 |
|-|-|-|
| `sorted(lst)` | **変更しない** (新しいリストを返す) | ソート済みリスト |
| `lst.sort()` | **元を変更する** | `None` |


````python
# sorted (元は変更されない)
a = [3, 1, 2]
b = sorted(a)
print('元:', a, '→ 戻り値:', b)

# sort (元が変わる)
a = [3, 1, 2]
ret = a.sort()
print('元:', a, '→ 戻り値:', ret)   # 戻り値は None !
````

## 3.6 リストの結合と繰り返し


````python
x = [1, 2, 3]
y = [4, 5, 6]

print(x + y)        # [1, 2, 3, 4, 5, 6]
print(x * 2)        # [1, 2, 3, 1, 2, 3]

# extend は + と同じく結合だが、元のリストを更新する
x.extend(y)
print(x)            # [1, 2, 3, 4, 5, 6]
````

## 3.7 2次元リスト (リストのリスト)

データ分析の "行と列" の表は、入れ子のリストで表せます。


````python
table = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9],
]

print(table[0])        # [1, 2, 3]    1行目
print(table[1][2])     # 6           2行目の3列目
print(len(table))      # 3           行数
print(len(table[0]))   # 3           列数
````

## 3.8 練習問題


````python
# Q1. len(sub) は何になる?
nums = [10, 20, 30, 40, 50, 60]
sub = nums[1:5]
print(sub, len(sub))
````

**Q1 解説**

`nums[1:5]` はインデックス 1,2,3,4 → `[20, 30, 40, 50]`、要素数 **4**。


````python
# Q2. 最終的な len(part) は?
part = [1, 2, 3]
part.append(4)
part.append(5)
part.pop(0)
print(part, len(part))
````

**Q2 解説**

1. `[1,2,3]` → append(4) → `[1,2,3,4]`
2. append(5) → `[1,2,3,4,5]`
3. `pop(0)` で先頭の 1 を削除 → `[2,3,4,5]`
4. `len(part)` = **4**


## まとめ

- リストは `[要素, ...]` で作成し、0 始まりのインデックスでアクセス
- スライス `[a:b]` の要素数は `b - a` 個 (end は含まない)
- `append`, `insert`, `pop`, `remove` で要素を変更
- スライスは元と別のリストを返す (コピー)
- `sorted` は新リスト, `sort` は元を変更
